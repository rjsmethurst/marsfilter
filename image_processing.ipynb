# Testing Image Manipulation with skimage for #marsfilter

```python
from ipywidgets import interact, interactive, fixed
from IPython.display import display
```

```python
import skimage
from skimage import data, filters, io
from scipy.ndimage import interpolation
from scipy.interpolate import interp1d
import numpy as np
from itertools import product
import matplotlib.image as mpimg
```

```python
i = mpimg.imread('/Users/becky/Pictures/IMG_0910_small.jpeg')
```

```python
def edit_image(image, sigma=0.1, r=1.0, g=0.465, b=0.25):
    new_image = filters.gaussian_filter(image, sigma=sigma, multichannel=True)
    new_image[:,:,0] = r*new_image[:,:,0]
    new_image[:,:,1] = g*new_image[:,:,1]
    new_image[:,:,2] = b*new_image[:,:,2]
    new_image = io.Image(new_image)
    return new_image
```

```python
lims = (0.0,1.0,0.01)
w = interactive(edit_image, image=fixed(i), sigma =(0, 10, 0.1), r=lims, g=lims, b=lims)
display(w)
```
